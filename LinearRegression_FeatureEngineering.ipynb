# Gaussian transformation
to check whether your data is normally distributed or not..If not then we can apply some mathematical equation and convert to Gaussian distribution

```python
plt.hist(csv_file['TV'])
```

```python
plt.hist(csv_file['Sales'])
```

```python
import scipy.stats as stat
import pylab
```

```python
def plot_data(df,features):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    csv_file[features].hist()
    plt.subplot(1,2,2)
    stat.probplot(csv_file[features], dist='norm', plot=pylab)
```

```python
plot_data(csv_file,'TV')
#if all these points come on a line then it is Gaussian distributed
```

# Logarithimic transforamtion to make GAUSSIAN distribution

```python
import numpy as np
csv_file['TV_log'] = np.log(csv_file['TV'])
plot_data(csv_file,'TV_log')

#we can not apply logarithimic transformation to make it Gaussian distribution
```

```python
#Reciprocal Trnasformation
csv_file['TV_reciprocal'] = 1/csv_file.TV
plot_data(csv_file, 'TV_reciprocal')
```

```python
#we can not take the above transformatin as it is not l
```

```python
#Square root transformation
```

```python
csv_file['TV_SQRT'] = csv_file.TV**(1/2)
plot_data(csv_file,'TV_SQRT' )
```

```python
#Exponential transformation
csv_file['TV_exp'] = csv_file.TV**(1/1.2)
plot_data(csv_file,'TV_exp' )
```

```python
csv_file['TV_exp'].parameters = stat.boxcox(csv_file['TV'])
```

```python
plot_data(csv_file, 'TV_exp')
```

```python
csv_file['TV_log'] = np.log1p(csv_file['TV'])
plot_data(csv_file, 'TV_log')
```
